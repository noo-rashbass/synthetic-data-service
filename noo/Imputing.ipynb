```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.impute import SimpleImputer
from datetime import date, timedelta
import pdb
```

```python
data = pd.read_csv('../isaFull.tsv', '\t')
data['Visit date [EUPATH_0000091]'] = pd.to_datetime(data['Visit date [EUPATH_0000091]'])
```

```python
pd.DataFrame(data.columns)
```

```python
minmax_time = [min(data['Visit date [EUPATH_0000091]']),\
               max(data['Visit date [EUPATH_0000091]'])]
```

```python
interesting_patient = data[data['Participant_Id']==1007].\
                        sort_values(['Visit date [EUPATH_0000091]']).reset_index(drop=True)
```

```python
minmax_time_patient = [min(interesting_patient['Visit date [EUPATH_0000091]']),\
               max(interesting_patient['Visit date [EUPATH_0000091]'])]
```

```python
interesting_patient[["Age at visit (years) [EUPATH_0000113]", "Temperature (C) [EUPATH_0000110]", "Visit date [EUPATH_0000091]"]]
```

```python
start_date = minmax_time[0]
start_date = minmax_time[0]
end_date = minmax_time[1]
delta = timedelta(days=1)
i = 0
df_with_nans = []
from_where = []
# pd.DataFrame(np.nan, index=[0],columns=data.columns)

while start_date <= end_date:
    if i < 38:
        if interesting_patient.loc[i, 'Visit date [EUPATH_0000091]'] == start_date:
            df_with_nans.append(interesting_patient.iloc[i])
            from_where.append('original')
            i += 1
        else:
            df_with_nans.append({'Visit date [EUPATH_0000091]':start_date})
            from_where.append('imputer')
    else:
        df_with_nans.append({'Visit date [EUPATH_0000091]':start_date})
        from_where.append('imputer')
    start_date += delta

df_with_nans = pd.DataFrame(df_with_nans)
df_with_nans['from_where'] = from_where
```

```python
selected_cols = df_with_nans[["Age at visit (years) [EUPATH_0000113]", 
              "Temperature (C) [EUPATH_0000110]"]]
selected_cols_arr = np.array(selected_cols)
imp = SimpleImputer(missing_values=np.nan, strategy='mean') 
imp.fit(selected_cols_arr)
impute = imp.transform(selected_cols_arr)
```

```python
imputed = pd.concat([pd.DataFrame(impute), df_with_nans[['Visit date [EUPATH_0000091]','from_where']]], axis=1)
imputed.columns = 'Age', 'Temp', 'Date', 'From'
imputed
```

```python
def plot_scatter(df,col):
    fig = go.Figure()
    fig.add_trace(go.Scatter(
                        x=df['Date'], 
                        y=df[col],
                        mode='markers',
                        name='lines',
                        marker=dict(
                            color=(
                                (df['From']=='imputer')
                            ).astype('int'),
                            colorscale=[[0, 'blue'], [1, 'green']]
                        )
                    ))
    fig.update_layout(title='Imputing {}'.format(col),
                   xaxis_title='Date',
                   yaxis_title='{}'.format(col))
    fig.show()
plot_scatter(imputed,'Age')
plot_scatter(imputed,'Temp')
```
