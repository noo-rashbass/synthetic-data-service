```python
import pandas as pd
import numpy as np

##import SIM I data
SIMI_SACT_TUMOUR=pd.read_table("simulacrum_release_v1/sim_sact_tumour.csv", sep=",")
```

```python
##randomise data on Tumour Id
SIMI_SACT_TUMOUR_50_tid = SIMI_SACT_TUMOUR.sample(frac =.50)
SIMI_SACT_TUMOUR_25_tid = SIMI_SACT_TUMOUR.sample(frac =.25)
SIMI_SACT_TUMOUR_10_tid = SIMI_SACT_TUMOUR.sample(frac =.10)
```

```python
##randomise data on Patient Id
SIMI_SACT_TUMOUR_pid = SIMI_SACT_TUMOUR.set_index('MERGED_PATIENT_ID')

##add number of tumour counts per patient Id
SIMI_SACT_TUMOUR_pid['counts'] = SIMI_SACT_TUMOUR_pid.groupby(['MERGED_PATIENT_ID']).size()
SIMI_SACT_TUMOUR_pid = SIMI_SACT_TUMOUR_pid.sort_index()

SIMI_SACT_TUMOUR_50_pid = SIMI_SACT_TUMOUR_pid.loc[np.random.choice(SIMI_SACT_TUMOUR_pid.index.unique(), int(SIMI_SACT_TUMOUR_pid.index.nunique()/2), replace=False)]
SIMI_SACT_TUMOUR_25_pid = SIMI_SACT_TUMOUR_pid.loc[np.random.choice(SIMI_SACT_TUMOUR_pid.index.unique(), int(SIMI_SACT_TUMOUR_pid.index.nunique()/4), replace=False)]
SIMI_SACT_TUMOUR_10_pid = SIMI_SACT_TUMOUR_pid.loc[np.random.choice(SIMI_SACT_TUMOUR_pid.index.unique(), int(SIMI_SACT_TUMOUR_pid.index.nunique()/10), replace=False)]
```

```python
##number of tumours of SIM I and samples randomised on Patient Id

def tumour_number (table_count, table, counts):

    table_count = pd.value_counts(table['counts']).to_frame().reset_index()
    table_count.columns=['n_tumours',counts]
    table_count[counts]=table_count[counts]/table_count["n_tumours"]
    
    return table_count
    
tumour_number_count=tumour_number('tumour_number_count',SIMI_SACT_TUMOUR_pid,"counts_total")
tumour_number_count_50_pid=tumour_number('tumour_number_count_50_pid',SIMI_SACT_TUMOUR_50_pid,"counts_0.50_pid")
tumour_number_count_25_pid=tumour_number('tumour_number_count_25_pid',SIMI_SACT_TUMOUR_25_pid,"counts_0.25_pid")
tumour_number_count_10_pid=tumour_number('tumour_number_count_10_pid',SIMI_SACT_TUMOUR_10_pid, "counts_0.10_pid")

tumour_freq_merge_pid=pd.merge(tumour_number_count,tumour_number_count_50_pid,on="n_tumours", how='outer')
tumour_freq_merge_pid=pd.merge(tumour_freq_merge_pid,tumour_number_count_25_pid,on="n_tumours", how='outer')
tumour_freq_merge_pid=pd.merge(tumour_freq_merge_pid,tumour_number_count_10_pid,on="n_tumours", how='outer')
tumour_freq_merge_pid=tumour_freq_merge_pid.fillna(0)

tumour_freq_merge_pid['counts_total_percentage']=tumour_freq_merge_pid['counts_total']/sum(tumour_freq_merge_pid['counts_total'])
tumour_freq_merge_pid['counts_0.50_pid_percentage']=tumour_freq_merge_pid['counts_0.50_pid']/sum(tumour_freq_merge_pid['counts_0.50_pid'])
tumour_freq_merge_pid['counts_0.25_pid_percentage']=tumour_freq_merge_pid['counts_0.25_pid']/sum(tumour_freq_merge_pid['counts_0.25_pid'])
tumour_freq_merge_pid['counts_0.10_pid_percentage']=tumour_freq_merge_pid['counts_0.10_pid']/sum(tumour_freq_merge_pid['counts_0.10_pid'])
```

```python
##number of tumours of SIM I and samples randomised on Tumour Id

def patient_index (table_new,table):
    table_new = table.set_index('MERGED_PATIENT_ID')
    table_new['counts']=table_new.groupby(['MERGED_PATIENT_ID']).size()
    table_new=table_new.sort_index()
    return table_new

SIMI_SACT_TUMOUR_50_tid2 = patient_index('SIMI_SACT_TUMOUR_50_tid2',SIMI_SACT_TUMOUR_50_tid)
SIMI_SACT_TUMOUR_25_tid2 = patient_index('SIMI_SACT_TUMOUR_25_tid2',SIMI_SACT_TUMOUR_25_tid)
SIMI_SACT_TUMOUR_10_tid2 = patient_index('SIMI_SACT_TUMOUR_25_tid2',SIMI_SACT_TUMOUR_10_tid)

tumour_number_count_50_tid=tumour_number('tumour_number_count_50_tid',SIMI_SACT_TUMOUR_50_tid2,"counts_0.50_tid")
tumour_number_count_25_tid=tumour_number('tumour_number_count_25_tid',SIMI_SACT_TUMOUR_25_tid2,"counts_0.25_tid")
tumour_number_count_10_tid=tumour_number('tumour_number_count_10_tid',SIMI_SACT_TUMOUR_10_tid2, "counts_0.10_tid")

tumour_freq_merge_tid=pd.merge(tumour_number_count,tumour_number_count_50_tid,on="n_tumours", how='outer')
tumour_freq_merge_tid=pd.merge(tumour_freq_merge_tid,tumour_number_count_25_tid,on="n_tumours", how='outer')
tumour_freq_merge_tid=pd.merge(tumour_freq_merge_tid,tumour_number_count_10_tid,on="n_tumours", how='outer')
tumour_freq_merge_tid=tumour_freq_merge_tid.fillna(0)

tumour_freq_merge_tid['counts_total_percentage']=tumour_freq_merge_tid['counts_total']/sum(tumour_freq_merge_tid['counts_total'])
tumour_freq_merge_tid['counts_0.50__tid_percentage']=tumour_freq_merge_tid['counts_0.50_tid']/sum(tumour_freq_merge_tid['counts_0.50_tid'])
tumour_freq_merge_tid['counts_0.25_tid_percentage']=tumour_freq_merge_tid['counts_0.25_tid']/sum(tumour_freq_merge_tid['counts_0.25_tid'])
tumour_freq_merge_tid['counts_0.10_tid_percentage']=tumour_freq_merge_tid['counts_0.10_tid']/sum(tumour_freq_merge_tid['counts_0.10_tid'])
```

```python
##frequency tables for SIM I and samples randomised on tumour Id

def frequency (table, table_freq, column, count):
    table_freq=pd.value_counts(table[column]).to_frame().reset_index()
    table_freq.columns=[column,count]
    return table_freq

SIMI_SACT_TUMOUR_tid_CONSULTANT_SPECIALITY_CODE_FREQ=frequency(SIMI_SACT_TUMOUR,'SIMI_SACT_TUMOUR_tid_CONSULTANT_SPECIALITY_CODE_FREQ','CONSULTANT_SPECIALITY_CODE','count_synth')
SIMI_SACT_TUMOUR_50_tid_CONSULTANT_SPECIALITY_CODE_FREQ=frequency(SIMI_SACT_TUMOUR_50_tid,'SIMI_SACT_TUMOUR_50_tid_CONSULTANT_SPECIALITY_CODE_FREQ','CONSULTANT_SPECIALITY_CODE','count_0.50_synth_tid')
SIMI_SACT_TUMOUR_25_tid_CONSULTANT_SPECIALITY_CODE_FREQ=frequency(SIMI_SACT_TUMOUR_25_tid,'SIMI_SACT_TUMOUR_25_tid_CONSULTANT_SPECIALITY_CODE_FREQ','CONSULTANT_SPECIALITY_CODE','count_0.25_synth_tid')
SIMI_SACT_TUMOUR_10_tid_CONSULTANT_SPECIALITY_CODE_FREQ=frequency(SIMI_SACT_TUMOUR_10_tid,'SIMI_SACT_TUMOUR_10_tid_CONSULTANT_SPECIALITY_CODE_FREQ','CONSULTANT_SPECIALITY_CODE','count_0.10_synth_tid')

SIMI_SACT_TUMOUR_tid_PRIMARY_DIAGNOSIS_FREQ=frequency(SIMI_SACT_TUMOUR,'SIMI_SACT_TUMOUR_tid_PRIMARY_DIAGNOSIS_FREQ','PRIMARY_DIAGNOSIS','count_synth')
SIMI_SACT_TUMOUR_50_tid_PRIMARY_DIAGNOSIS_FREQ=frequency(SIMI_SACT_TUMOUR_50_tid,'SIMI_SACT_TUMOUR_50_tid_PRIMARY_DIAGNOSIS_FREQ','PRIMARY_DIAGNOSIS','count_0.50_synth_tid')
SIMI_SACT_TUMOUR_25_tid_PRIMARY_DIAGNOSIS_FREQ=frequency(SIMI_SACT_TUMOUR_25_tid,'SIMI_SACT_TUMOUR_25_tid_PRIMARY_DIAGNOSIS_FREQ','PRIMARY_DIAGNOSIS','count_0.25_synth_tid')
SIMI_SACT_TUMOUR_10_tid_PRIMARY_DIAGNOSIS_FREQ=frequency(SIMI_SACT_TUMOUR_10_tid,'SIMI_SACT_TUMOUR_10_tid_PRIMARY_DIAGNOSIS_FREQ','PRIMARY_DIAGNOSIS','count_0.10_synth_tid')

SIMI_SACT_TUMOUR_tid_MORPHOLOGY_CLEAN_FREQ=frequency(SIMI_SACT_TUMOUR,'SIMI_SACT_TUMOUR_tid_MORPHOLOGY_CLEAN_FREQ','MORPHOLOGY_CLEAN','count_synth')
SIMI_SACT_TUMOUR_50_tid_MORPHOLOGY_CLEAN_FREQ=frequency(SIMI_SACT_TUMOUR_50_tid,'SIMI_SACT_TUMOUR_50_tid_MORPHOLOGY_CLEAN_FREQ','MORPHOLOGY_CLEAN','count_0.50_synth_tid')
SIMI_SACT_TUMOUR_25_tid_MORPHOLOGY_CLEAN_FREQ=frequency(SIMI_SACT_TUMOUR_25_tid,'SIMI_SACT_TUMOUR_25_tid_MORPHOLOGY_CLEAN_FREQ','MORPHOLOGY_CLEAN','count_0.25_synth_tid')
SIMI_SACT_TUMOUR_10_tid_MORPHOLOGY_CLEAN_FREQ=frequency(SIMI_SACT_TUMOUR_10_tid,'SIMI_SACT_TUMOUR_10_tid_MORPHOLOGY_CLEAN_FREQ','MORPHOLOGY_CLEAN','count_0.10_synth_tid')

##merging the frequency tables

def frequency_merged (table_freq_merge, table_freq, table50_freq, table25_freq, table10_freq, column):
    table_freq_merge=pd.merge(table_freq,table50_freq,on=column, how='outer')
    table_freq_merge=pd.merge(table_freq_merge,table25_freq,on=column, how='outer')
    table_freq_merge=pd.merge(table_freq_merge,table10_freq,on=column, how='outer')
    table_freq_merge.sort_values(by='count_synth',ascending=False)
    
    return table_freq_merge

CONSULTANT_SPECIALITY_CODE_tid_FREQ_merge=frequency_merged('CONSULTANT_SPECIALITY_CODE_tid_FREQ_merge',SIMI_SACT_TUMOUR_tid_CONSULTANT_SPECIALITY_CODE_FREQ,SIMI_SACT_TUMOUR_50_tid_CONSULTANT_SPECIALITY_CODE_FREQ,SIMI_SACT_TUMOUR_25_tid_CONSULTANT_SPECIALITY_CODE_FREQ,SIMI_SACT_TUMOUR_10_tid_CONSULTANT_SPECIALITY_CODE_FREQ,'CONSULTANT_SPECIALITY_CODE')
CONSULTANT_SPECIALITY_CODE_tid_FREQ_merge=CONSULTANT_SPECIALITY_CODE_tid_FREQ_merge.fillna(0)
PRIMARY_DIAGNOSIS_FREQ_tid_merge=frequency_merged('PRIMARY_DIAGNOSIS_FREQ_tid_merge',SIMI_SACT_TUMOUR_tid_PRIMARY_DIAGNOSIS_FREQ,SIMI_SACT_TUMOUR_50_tid_PRIMARY_DIAGNOSIS_FREQ,SIMI_SACT_TUMOUR_25_tid_PRIMARY_DIAGNOSIS_FREQ,SIMI_SACT_TUMOUR_10_tid_PRIMARY_DIAGNOSIS_FREQ,'PRIMARY_DIAGNOSIS')
PRIMARY_DIAGNOSIS_FREQ_tid_merge=PRIMARY_DIAGNOSIS_FREQ_tid_merge.fillna(0)
MORPHOLOGY_CLEAN_FREQ_tid_merge=frequency_merged('MORPHOLOGY_CLEAN_FREQ_tid_merge',SIMI_SACT_TUMOUR_tid_MORPHOLOGY_CLEAN_FREQ,SIMI_SACT_TUMOUR_50_tid_MORPHOLOGY_CLEAN_FREQ,SIMI_SACT_TUMOUR_25_tid_MORPHOLOGY_CLEAN_FREQ,SIMI_SACT_TUMOUR_10_tid_MORPHOLOGY_CLEAN_FREQ,'MORPHOLOGY_CLEAN')
MORPHOLOGY_CLEAN_FREQ_tid_merge=MORPHOLOGY_CLEAN_FREQ_tid_merge.fillna(0)
```

```python
##frequency tables for SIM I and samples randomised on patient Id

SIMI_SACT_TUMOUR_pid_CONSULTANT_SPECIALITY_CODE_FREQ=frequency(SIMI_SACT_TUMOUR_pid,'SIMI_SACT_TUMOUR_pid_CONSULTANT_SPECIALITY_CODE_FREQ','CONSULTANT_SPECIALITY_CODE','count_synth')
SIMI_SACT_TUMOUR_50_pid_CONSULTANT_SPECIALITY_CODE_FREQ=frequency(SIMI_SACT_TUMOUR_50_pid,'SIMI_SACT_TUMOUR_50_pid_CONSULTANT_SPECIALITY_CODE_FREQ','CONSULTANT_SPECIALITY_CODE','count_0.50_synth_pid')
SIMI_SACT_TUMOUR_25_pid_CONSULTANT_SPECIALITY_CODE_FREQ=frequency(SIMI_SACT_TUMOUR_25_pid,'SIMI_SACT_TUMOUR_25_pid_CONSULTANT_SPECIALITY_CODE_FREQ','CONSULTANT_SPECIALITY_CODE','count_0.25_synth_pid')
SIMI_SACT_TUMOUR_10_pid_CONSULTANT_SPECIALITY_CODE_FREQ=frequency(SIMI_SACT_TUMOUR_10_pid,'SIMI_SACT_TUMOUR_10_pid_CONSULTANT_SPECIALITY_CODE_FREQ','CONSULTANT_SPECIALITY_CODE','count_0.10_synth_pid')

SIMI_SACT_TUMOUR_pid_PRIMARY_DIAGNOSIS_FREQ=frequency(SIMI_SACT_TUMOUR_pid,'SIMI_SACT_TUMOUR_tid_PRIMARY_DIAGNOSIS_FREQ','PRIMARY_DIAGNOSIS','count_synth')
SIMI_SACT_TUMOUR_50_pid_PRIMARY_DIAGNOSIS_FREQ=frequency(SIMI_SACT_TUMOUR_50_pid,'SIMI_SACT_TUMOUR_50_pid_PRIMARY_DIAGNOSIS_FREQ','PRIMARY_DIAGNOSIS','count_0.50_synth_pid')
SIMI_SACT_TUMOUR_25_pid_PRIMARY_DIAGNOSIS_FREQ=frequency(SIMI_SACT_TUMOUR_25_pid,'SIMI_SACT_TUMOUR_25_pid_PRIMARY_DIAGNOSIS_FREQ','PRIMARY_DIAGNOSIS','count_0.25_synth_pid')
SIMI_SACT_TUMOUR_10_pid_PRIMARY_DIAGNOSIS_FREQ=frequency(SIMI_SACT_TUMOUR_10_pid,'SIMI_SACT_TUMOUR_10_pid_PRIMARY_DIAGNOSIS_FREQ','PRIMARY_DIAGNOSIS','count_0.10_synth_pid')

SIMI_SACT_TUMOUR_pid_MORPHOLOGY_CLEAN_FREQ=frequency(SIMI_SACT_TUMOUR_pid,'SIMI_SACT_TUMOUR_pid_MORPHOLOGY_CLEAN_FREQ','MORPHOLOGY_CLEAN','count_synth')
SIMI_SACT_TUMOUR_50_pid_MORPHOLOGY_CLEAN_FREQ=frequency(SIMI_SACT_TUMOUR_50_pid,'SIMI_SACT_TUMOUR_50_pid_MORPHOLOGY_CLEAN_FREQ','MORPHOLOGY_CLEAN','count_0.50_synth_pid')
SIMI_SACT_TUMOUR_25_pid_MORPHOLOGY_CLEAN_FREQ=frequency(SIMI_SACT_TUMOUR_25_pid,'SIMI_SACT_TUMOUR_25_pid_MORPHOLOGY_CLEAN_FREQ','MORPHOLOGY_CLEAN','count_0.25_synth_pid')
SIMI_SACT_TUMOUR_10_pid_MORPHOLOGY_CLEAN_FREQ=frequency(SIMI_SACT_TUMOUR_10_pid,'SIMI_SACT_TUMOUR_10_pid_MORPHOLOGY_CLEAN_FREQ','MORPHOLOGY_CLEAN','count_0.10_synth_pid')

##merging the frequency tables

CONSULTANT_SPECIALITY_CODE_pid_FREQ_merge=frequency_merged('CONSULTANT_SPECIALITY_CODE_pid_FREQ_merge',SIMI_SACT_TUMOUR_pid_CONSULTANT_SPECIALITY_CODE_FREQ,SIMI_SACT_TUMOUR_50_pid_CONSULTANT_SPECIALITY_CODE_FREQ,SIMI_SACT_TUMOUR_25_pid_CONSULTANT_SPECIALITY_CODE_FREQ,SIMI_SACT_TUMOUR_10_pid_CONSULTANT_SPECIALITY_CODE_FREQ,'CONSULTANT_SPECIALITY_CODE')
CONSULTANT_SPECIALITY_CODE_pid_FREQ_merge=CONSULTANT_SPECIALITY_CODE_pid_FREQ_merge.fillna(0)
PRIMARY_DIAGNOSIS_FREQ_pid_merge=frequency_merged('PRIMARY_DIAGNOSIS_FREQ_pid_merge',SIMI_SACT_TUMOUR_pid_PRIMARY_DIAGNOSIS_FREQ,SIMI_SACT_TUMOUR_50_pid_PRIMARY_DIAGNOSIS_FREQ,SIMI_SACT_TUMOUR_25_pid_PRIMARY_DIAGNOSIS_FREQ,SIMI_SACT_TUMOUR_10_pid_PRIMARY_DIAGNOSIS_FREQ,'PRIMARY_DIAGNOSIS')
PRIMARY_DIAGNOSIS_FREQ_pid_merge=PRIMARY_DIAGNOSIS_FREQ_pid_merge.fillna(0)
MORPHOLOGY_CLEAN_FREQ_pid_merge=frequency_merged('MORPHOLOGY_CLEAN_FREQ_pid_merge',SIMI_SACT_TUMOUR_pid_MORPHOLOGY_CLEAN_FREQ,SIMI_SACT_TUMOUR_50_pid_MORPHOLOGY_CLEAN_FREQ,SIMI_SACT_TUMOUR_25_pid_MORPHOLOGY_CLEAN_FREQ,SIMI_SACT_TUMOUR_10_pid_MORPHOLOGY_CLEAN_FREQ,'MORPHOLOGY_CLEAN')
MORPHOLOGY_CLEAN_FREQ_pid_merge=MORPHOLOGY_CLEAN_FREQ_pid_merge.fillna(0)
```
